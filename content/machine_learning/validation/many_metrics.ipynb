It's easy to get overwhelmed at the sheer number of accuracy metrics you can calculate from a simple Confusion Matrix

```python
%pylab inline
from IPython.display import Image

Image('images/lots_of_metrics.jpg')
```

```python
%pylab inline

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

X, y = make_classification(1000)
model = LogisticRegression()
model.fit(X, y)

y_pred_probs = model.predict_proba(X)[:, 1]
y_pred = model.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
from helpers import print_cm

print_cm(confusion_matrix(y, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y, y_pred)
```

```python
from helpers import plot_multiple_thresholds

plot_multiple_thresholds(y, y_pred_probs)
```

### Downsampled Class

```python
from helpers import downsample_class

X, y = make_classification(100000)
X, y = downsample_class(X, y)

model = LogisticRegression()
model.fit(X, y)

y_pred_probs = model.predict_proba(X)[:, 1]
y_pred = model.predict(X)
```
